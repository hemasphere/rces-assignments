## 1 Creating and Manipulating Arrays

```python
import numpy as np
from matplotlib import pyplot as plt
```

### 1.1 Create two 2D arrays representing coordinates x, y on the cartesian plan

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
xx, yy = np.meshgrid(x, y)
```

### 1.2 Visualize each 2D array using pcolormesh

```python
plt.pcolormesh(xx);
```

```python
plt.pcolormesh(yy);
```

### 1.3 From your cartesiam coordinates, create polar coordinates 𝑟 and  𝜑

```python
r = np.sqrt(xx**2 + yy**2)
```

```python
𝜑 = np.arctan2(yy, xx)
```

### 1.4 Visualize r and 𝜑 on the 2D x/y plane using pcolormesh

```python
plt.pcolormesh(xx, yy, r);
```

```python
plt.pcolormesh(𝜑);
```

### 1.5 Calculate the quantity f = cos2(4𝑟)+sin2(4𝜑)

```python
f = (np.cos(4*r))**2 + (np.sin(4*𝜑))**2
```

```python
plt.pcolormesh(f);
```

### 1.6 Plot the mean of f with respect to the x axis

```python
plt.plot(y, np.mean(f, 0));
```

### 1.7 Plot the mean of f with respect to the y axis

```python
plt.plot(x, np.mean(f, 1));
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

## 1) Use Pandas’ read_csv function directly on this URL to open it as a DataFrame¶

```python
datafile = "http://www.ldeo.columbia.edu/~rpa/usgs_earthquakes_2014.csv"
df = pd.read_csv(datafile)
df.head()
```

```python
df.info()
```

## 2) Re-read the data in such a way that all date columns are identified as dates and the earthquake ID is used as the index

```python
df = pd.read_csv(datafile, parse_dates=['time', 'updated'], index_col=['id'])
```

```python
df.info()
```

```python
df.head()
```

## 3) Use describe to get the basic statistics of all the columns

## 4) Use nlargest to get the top 20 earthquakes by magnitude

```python
df.nlargest(20, 'mag')
```

## 5) Extract the state or country using Pandas text data functions

```python
country = df['place'].str.split(", ").str.get(1)
df['country'] = country
del df['place']
```

```python
df
```

## 7) Create a filtered dataset that only has earthquakes of magnitude 4 or larger and

```python
filtered = df[(df.type == 'earthquake') & (df.mag >= 4)]
filtered
```

## 8) Using the filtered dataset (magnitude > 4), count the number of earthquakes in each country/state. Make a bar chart of this number for the top 5 locations with the most earthquakes

```python
top_five_largest = filtered.pivot_table(columns = ['country'], aggfunc = 'size').nlargest(5)
top_five_largest
```

```python
top_five_largest.nlargest(5).plot(kind = "bar")
```

## 9) Make a histogram the distribution of the Earthquake magnitudes

```python
#unfiltered 
unfiltered = df.mag.hist(bins = np.arange(-1, 8.4, 0.1), log = True)
unfiltered
```

```python
#filtered
f1 = filtered.mag.hist(bins = np.arange(4, 8.4, 0.1), log = True)
f1
```

## 10) Visualize the locations of earthquakes by making a scatterplot of their latitude and longitude

```python
fig, (df, filtered) = plt.subplots(ncols = 2)
df.plot.scatter(x = 'latitude', y = 'longitude', c = 'red')
filtered.plot.scatter(x = 'latitude', y = 'longitude', c = 'blue')
```

```python
fig, ax = plt.subplots(ncols = 2, figsize = (18, 6))
df.plot.scatter(x = 'longitude', y = 'latitude', c='mag', cmap='magma', s=2, vmin=0, vmax=8, ax=ax[0])
ax[0].grid()
filtered.plot.scatter(x = 'longitude', y = 'latitude', c='mag', cmap='magma', s=2, vmin=0, vmax=8, ax=ax[1])
ax[1].grid()
```
